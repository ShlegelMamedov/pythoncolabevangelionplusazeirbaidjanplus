```python
# 🧮 Простой калькулятор
# Авторы: Шлегель Матвей, Мамедов Рамин | Группа 2-ИС33

def calculator(num1, num2, operator):
    """
    Выполняет базовую математическую операцию.

    Args:
        num1 (float): Первое число
        num2 (float): Второе число
        operator (str): Оператор (+, -, *, /)

    Returns:
        Результат операции или сообщение об ошибке
    """
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "❌ Ошибка: деление на ноль!"
        return num1 / num2
    else:
        return "❌ Ошибка: недопустимый оператор!"

# === Основной код ===
print("🧮 Калькулятор от Шлегеля и Мамедова")
print("-" * 40)

try:
    # Ввод данных с обработкой ошибок
    num1 = float(input("Введите первое число: "))
    operator = input("Введите оператор (+, -, *, /): ").strip()
    num2 = float(input("Введите второе число: "))

    # Вычисление и вывод результата
    result = calculator(num1, num2, operator)
    print(f"\n✅ Результат: {num1} {operator} {num2} = {result}")

except ValueError:
    print("❌ Ошибка: введите корректные числа!")
```

```python
# 🔐 Генератор безопасных паролей
# Авторы: Шлегель Матвей, Мамедов Рамин | Группа 2-ИС33

import random
import string

def check_password_strength(password):
    """
    Оценивает сложность пароля.

    Returns:
        str: Уровень сложности (Слабый/Средний/Надёжный)
    """
    score = 0
    if len(password) >= 8:
        score += 1
    if len(password) >= 12:
        score += 1
    if any(c.islower() for c in password):
        score += 1
    if any(c.isupper() for c in password):
        score += 1
    if any(c.isdigit() for c in password):
        score += 1
    if any(c in string.punctuation for c in password):
        score += 1

    if score <= 2:
        return "🔴 Слабый"
    elif score <= 4:
        return "🟡 Средний"
    else:
        return "🟢 Надёжный"

def generate_password(length=12, use_letters=True, use_digits=True, use_special=True):
    """
    Генерирует безопасный пароль.

    Args:
        length (int): Длина пароля
        use_letters (bool): Использовать буквы
        use_digits (bool): Использовать цифры
        use_special (bool): Использовать спецсимволы

    Returns:
        str: Сгенерированный пароль
    """
    characters = ""
    if use_letters:
        characters += string.ascii_letters  # a-zA-Z
    if use_digits:
        characters += string.digits          # 0-9
    if use_special:
        characters += string.punctuation     # !@#$%^&*()

    if not characters:
        return "❌ Ошибка: выберите хотя бы один тип символов!"

    # Генерация пароля
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def generate_multiple_passwords(count=3, **kwargs):
    """Генерирует несколько паролей."""
    print(f"\n🔐 Сгенерировано {count} паролей:")
    print("-" * 50)
    for i in range(count):
        pwd = generate_password(**kwargs)
        strength = check_password_strength(pwd)
        print(f"{i+1}. {pwd} [{strength}]")

# === Основной код ===
print("🔐 Генератор паролей от Шлегеля и Мамедова")
print("=" * 50)

# Настройки
try:
    length = int(input("\n📏 Длина пароля (мин. 4): ") or 12)
    length = max(4, length)  # Минимальная длина

    use_letters = input("🔤 Использовать буквы? (да/нет): ").lower() != 'нет'
    use_digits = input("🔢 Использовать цифры? (да/нет): ").lower() != 'нет'
    use_special = input("✨ Использовать спецсимволы? (да/нет): ").lower() != 'нет'

    count = int(input("📦 Сколько паролей сгенерировать? (1-10): ") or 3)
    count = max(1, min(10, count))  # Ограничение 1-10

    # Генерация и вывод
    generate_multiple_passwords(
        count=count,
        length=length,
        use_letters=use_letters,
        use_digits=use_digits,
        use_special=use_special
    )

    # Совет по безопасности
    print("\n💡 Совет: сохраните пароль в надёжном менеджере паролей!")

except ValueError:
    print("❌ Ошибка: введите корректные числовые значения!")
```
